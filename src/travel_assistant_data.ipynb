```python
import os
import pandas as pd
```

```python
# get the list of data files from resource dir
data_files = os.listdir("../Resources/Reviews")
data_files
```

```python
# Hotel Reviews
h_1 = pd.read_csv("../Resources/Reviews/hotel_reviews_1.csv")
h_2 = pd.read_csv("../Resources/Reviews/hotel_reviews_2.csv")
h_3 = pd.read_csv("../Resources/Reviews/hotel_reviews_3.csv")

h_1.drop(columns="reviews.dateAdded", axis=1, inplace=True)
h_2.drop(columns="reviews.dateAdded", axis=1, inplace=True)

if set(h_1.columns) == set(h_2.columns) == set(h_3.columns): 
    print("all have same columns")
    print(h_1.columns)
else:
    print("need to adjust columns")
```

```python
hotel_reviews_df = pd.concat([h_1, h_2, h_3]).reset_index(drop=True)
hotel_reviews_df.info()
```

```python
# drop the duplicate rows 
hotel_reviews_df.drop_duplicates(keep='first', inplace=True)
hotel_reviews_df.reset_index(drop=True, inplace=True)
hotel_reviews_df.info()
```

```python
hotel_reviews_df[:5000].to_csv("../Resources/Reviews/Cleaned/hotel_reviews_1.csv", index=False)
hotel_reviews_df[5001:10000].to_csv("../Resources/Reviews/Cleaned/hotel_reviews_2.csv", index=False)
hotel_reviews_df[10001:15000].to_csv("../Resources/Reviews/Cleaned/hotel_reviews_3.csv", index=False)
hotel_reviews_df[15001:20000].to_csv("../Resources/Reviews/Cleaned/hotel_reviews_4.csv", index=False)
hotel_reviews_df[20001:].to_csv("../Resources/Reviews/Cleaned/hotel_reviews_5.csv", index=False)

```
